```python
# Import the main plotting object
import matplotlib.pyplot as plt

# Let's plot a simple graph
plt.plot([1,2,3,4,5,6], [2,1,3,1,4,1])
```

```python
# Let's generate a plot with more than one line
# We'll use the numpy library to generate some random numbers
import numpy as np

for i in range(4):
    plt.plot([1,2,3,4,5,6], np.random.normal(size=6))
```

```python
# Barchart
plt.bar([1,2,3,4,5,6], np.random.normal(size=6))
```

```python
# Scatterplot
plt.scatter([1,2,3,4,5,6], np.random.normal(size=6))
```

```python
# Histogram (with some extra fun code)
data = np.random.normal(size=1000)
counts, bins, patches = plt.hist(data)
print "maximum: {} in bin {:.3f}".format(
    max(counts),
    bins[np.argmax(counts)], # get bin where the count is maximum
)
```

# Multiple panels

Specifying subfigures or multiple panels in one figure is somewhat more complicated.
The most powerful and useful, but slightly difficult way is with `GridSpec`.
Here, you have fine-grained control over the layout and you can make many figures possible.
A nice explanation can be found here:

http://worksofscience.net/matplotlib/gridspec

Here's an example for two plots.

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=[8,3])
# create a grid with one row and two columns
gs = GridSpec(1,2)

# select the first row and first column
ax1 = fig.add_subplot(gs[0,0])
plt.plot([1,2,3],[3,2,1])

# select the first row and second column
ax2 = fig.add_subplot(gs[0,1])
plt.scatter([1,2,3],[3,2,1], color='red')
```

## More fancy, three plots that have different sizes.

```python
fig = plt.figure(figsize=[8,3])

# create a grid with two rows and two columns
gs = GridSpec(2,2)

# select all rows and first column
ax1 = fig.add_subplot(gs[0:2,0])
plt.plot([1,2,3],[3,2,1])

# select the first row and second column
ax2 = fig.add_subplot(gs[0,1])
plt.scatter([1,2,3],[3,2,1], color='red')

# select the second row and second column
ax3 = fig.add_subplot(gs[1,1])
plt.scatter([1,2,3],[3,2,1], color='green')
```

## Illustration of some extra options

```python
fig = plt.figure(figsize=[8,4])
# create a grid with two rows and two columns
gs = GridSpec(2,2)

# select all rows and first column
ax1 = fig.add_subplot(gs[0:2,0])
plt.plot([1,2,3],[3,2,1])

# select the first row and second column
ax2 = fig.add_subplot(gs[0,1])
plt.scatter([1,2,3],[3,2,1], color='red')

# select the second row and second column
ax3 = fig.add_subplot(gs[1,1])
plt.scatter([1,2,3],[3,2,1], color='green')

# you can still work with the plots you've plotted earlier
ax2.set_title("second plot")

# or like this
# set the current axis
plt.sca(ax1)
# now all plotting commands will use that axis
plt.xlabel("x axis")
plt.title("first plot")

# this is the most simple way to have enough space for labels and such
plt.tight_layout()
```
